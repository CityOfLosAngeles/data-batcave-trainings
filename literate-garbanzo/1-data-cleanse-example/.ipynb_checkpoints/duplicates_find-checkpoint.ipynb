```python
# Import libraries
from __future__ import print_function
import sys
from datetime import datetime
import time
import pandas as pd
```

```python
#print("Start time: " + str(datetime.now()))
```

```python
file_in = "Resources/SFPD_CrimeIncidents_2018.csv"
file_out_dupes ="Resources/SF_Police_Department_Incidents_dupes.csv"
# read in CSV
df = pd.read_csv(file_in)
df.head()
```

```python
# Print length of dataframe
print(len(df))
```

```python
# assign field name
field_name = 'IncidntNum'
```

```python
key_field = set(df[field_name].values)
header = True
```

```python
# concatenate values in instances where field_name value > 1
data = pd.concat(g for _, g in df.groupby(field_name) if len(g) > 1)
if len(data) > 1:
    print('Found duplicates')
    data.to_csv(file_out_dupes, mode='w', header=header, index=False)
else:
    pass
```
